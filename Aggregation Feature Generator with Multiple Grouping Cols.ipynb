```python
import datatable as dt
from datatable import join, by, f, min, max, mean, sd
import pandas as pd
from itertools import permutations, combinations
from random import sample
```

```python
X = dt.fread("churn_test.csv")
XP = X.to_pandas()
```

```python
uniques = X.nunique().to_list()
nrows = X.nrows
ncols = X.ncols
```

```python
X_num = X[:, [float, int]]
X_cat = X[:, [str]]
```

```python
X_num
type(X_num.names)
```

```python
strlist = []
[strlist.append(s) for s in X_cat.names]

strlist.remove("customer_id")
intlist = []
[intlist.append(i) for i in X_num.names]
    
print(strlist)
print(intlist)
agglist = ['mean','max']
```

```python
# Generate combination grouping
# gglist at the end has single col + combinations of multiple cat columns
#
max_len_groups = 4 # Maximum of 4 col group bys
gglist = []

# Include all individual cols by default
[gglist.append(j) for j in strlist]

# Create combinations (ordered permutations)
# Assumption, cat column order is somewhat hiearchical 
tmp_gg_list = []
for j in range(1, max_len_groups):
    comb = combinations(strlist,j+1) 
    comb_list = list(comb)
    for r in comb_list:
        if (len(r) <= max_len_groups):
            tmp_gg_list.append(r) 
            
[gglist.append(x) for x in tmp_gg_list]
```

```python
gglist
```

```python
len(gglist)*len(intlist)*len(agglist)
```

```python
groups = 0

for i in gglist: # Walk through Group By cols, 1 Combination at a time
    y = []
    # y - holds the BY clause for data table GROUP BY
    # agg_y_colname - holds Column name to go into data table. Otherwise default will be C0
    if (isinstance(i,tuple)):
        y = ','.join("dt.f."+g for g in i)     
        agg_y_colname = '_'.join(g for g in i)
    else:
        y = "dt.f."+i
        agg_y_colname = i  

    for j in range(0,len(intlist)): # Walk through Numeric cols 1 by 1
        x = intlist[j]
        for z in range(0,len(agglist)): # Walk through Aggregates 1 by 1
            agg = agglist[z]
            # Final column name a mix of GROUP BY column name, Numeric Column and Aggregator
            agg_colname = "GroupBy_"+agg_y_colname.upper()+"_"+agg+"_of_"+x.upper()
            # FINAL DATA TABLE GROUP BY CLAUSE
            group_by = "X[:, dt."+agg+"(dt.f."+x+"),by("+y+")]"
            groups +=1
            F = eval(group_by) # DATA TABLE GROUP BY EXECUTION
            F.names = {'C0':agg_colname}
            print(str(groups)+":  "+agg_colname)
            c = F[:,1].countna()
            # Add to features only if there are no NAs in the cols
            if (c[0,0] == 0):  
                FP = F.to_pandas()
                XP = pd.merge(XP, FP, how='inner')       
            
```
